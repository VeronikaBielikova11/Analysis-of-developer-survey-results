```python
import pandas as pd
public_res = pd.read_csv("survey_results_public.csv")
schema_res = pd.read_csv("survey_results_schema.csv")
pd.set_option('display.max_columns', None)
```

```python
# Скільки респондентів пройшли опитування?
totalCount = public_res['ResponseId'].nunique()
print(f"Пройшли опитування: {totalCount} респондентів")
```

```python
# Скільки респондентів відповіли на всі запитання?
all_quest = set(schema_res['qname'])
quest_in_public = set(public_res.columns)
intersecq = list(all_quest.intersection(quest_in_public))
answeredq = public_res[intersecq].dropna().shape[0]
print(f"Відповіли на всі запитання: {answeredq} респондентів")
```

```python
# Які значення мір центральної тенденції для досвіду (WorkExp) респондентів?
meanWorkExp = public_res["WorkExp"].mean().round(2)
medianWorkExp = public_res["WorkExp"].median()
modeWorkExp = public_res["WorkExp"].mode().iloc[0]
print(f"Mean by WorkExp: {meanWorkExp}\nMedian by WorkExp: {medianWorkExp}\nMode by WorkExp: {modeWorkExp}")
```

```python
# Скільки респондентів працює віддалено?
remote = public_res[ 
    public_res['RemoteWork'].str.contains('remote', case=False, na=False) 
]
remote_count = remote.shape[0]
print('Count remote workers:', remote_count)
```

```python
# Скільки респондентів навчалося програмувати за допомогою онлайн курсів?
courses = public_res[
public_res['LearnCode'].str.contains('Online courses', case = False, na=False)
]
count_workers_inOnlineCourse = courses.shape[0]
print('Навчалися програмувати за допомогою онлайн курсів:', count_workers_inOnlineCourse)
```

```python
# Який відсоток респондентів програмує на Python?
learn_python = public_res[
public_res['LanguageHaveWorkedWith'].str.contains('Python', case = False, na=False)
]
count_leaners = learn_python.shape[0]

pythonRespondents = round((count_leaners/totalCount * 100),2)
print(f'Відсоток респондентів, які програмують на Python: {pythonRespondents}')
```

```python
# Серед респондентів що програмують на Python в групуванні по країнам, яка середня та медіанна сума компенсації (ConvertedCompYearly) в кожній країні?
python_comp_by_country = learn_python.groupby('Country')['ConvertedCompYearly'].agg(
    Mean_comp = 'mean',
    Median_comp = 'median'
).reset_index()

python_comp_by_country
```

```python
top_5_Countries = python_comp_by_country.sort_values(by='Median_comp', ascending= False).head(5)
print(', '.join(top_5_Countries['Country'].tolist()))
```
